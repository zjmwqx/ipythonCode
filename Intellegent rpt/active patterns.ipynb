```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import regex
import numpy as np
import matplotlib.pylab as plt
from scipy.stats.mstats import mquantiles
```

```python
patterns = pd.read_csv('sentiment_patterns.txt',delimiter='\t',index_col='reportID')
```

```python
patterns = patterns.fillna('')
```

```python
?? CountVectorizer
```

```python
def build_dtm(records):
    dtm_builder = CountVectorizer(analyzer=lambda x: x.split(',') if len(x)>0 else [])
    X = dtm_builder.fit_transform(records)
    patterns = sorted(dtm_builder.vocabulary_.keys())
    return patterns, X
```

```python
pos_patterns, pos_dtm = build_dtm(patterns.posTerms)
neg_patterns, neg_dtm = build_dtm(patterns.negTerms)
```

```python
def build_df(patterns, dtm, orientation):
    tf = np.squeeze(np.asarray(dtm.sum(axis=0)))
    df = np.squeeze(np.asarray((dtm>0).sum(axis=0)))
    return pd.DataFrame({'pattern':patterns,'tf':tf,'df':df,'orientation':orientation})
```

```python
pattern_stats = pd.concat([build_df(pos_patterns, pos_dtm,'pos'),build_df(neg_patterns, neg_dtm, 'neg')],ignore_index=True)
```

```python
pattern_stats['ngram'] = pattern_stats.pattern.apply(lambda x: 1+len(regex.findall('\|',x)))
```

```python
pattern_stats.sort('tf',inplace=True,ascending=False)
```

```python
print('number of documents: %d' % patterns.shape[0])
print('number of active patterns: %d' % pattern_stats.shape[0])
print('number of active pos patterns: %d' % len(pos_patterns))
print('number of active neg patterns: %d' % len(neg_patterns))
```

```python
n_active_pattern = pattern_stats.shape[0]
n_document = patterns.shape[0]
```

```python
pattern_stats.to_csv('pattern_stats.csv',index=False)
```
