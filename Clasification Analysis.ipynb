```python
import pandas as pd
import numpy as np
from matplotlib import pylab as plt
import pymongo
```

```python
matplotlib.rcParams['font.family'] = 'WenQuanYi Zen Hei'
```

```python
con = pymongo.Connection('nosql01-dev.datayes.com', port=27017)
reportsInfo = con.reportQA.reportsInfo
```

```python
cursor = reportsInfo.find({'qaUpdateTime': {"$gte": "2013-12-20"}})
```

```python
rpt_fields = ['title', 'pat', 'patterns','alg1','alg2','qaCategory']
result = pd.DataFrame(list(cursor), columns=rpt_fields)
```

```python
result.alg1 = result.alg1.map(lambda x:tuple(x))
result.alg2 = result.alg2.map(lambda x:tuple(x))
result.pat = result.pat.map(lambda x:tuple(x))
result.qaCategory = result.qaCategory.map(lambda x: tuple(x))
result.patterns = result.patterns.map(lambda x: tuple(x))
```

s1:计算准确率：

```python
def equality(temp):
    return temp.apply(lambda x : x['qaCategory']==x,axis =1)
```

```python
def isUnCategory(temp):
    return temp.apply(lambda x : x==(u'未分类',),axis=1)
```

```python
def calHitRate(temp):
    return ((~isUnCategory(temp)) & equality(temp)).sum()/temp[~isUnCategory(temp)].count()
```

```python
result[isUnCategory(result)].count()
```

```python
calHitRate(result)#计算准确率，刨去未分类
```

```python
acc = result.groupby('pat').apply(calHitRate)
```

```python
acc = acc.sort()
```

```python
acc.shape
```

```python
acc.sort('pat',ascending=False).ix[1:,['pat','qaCategory','alg1','alg2']].plot(kind='barh',figsize=(10,20))
```
