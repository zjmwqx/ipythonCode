```python
path = 'usagov_bitly_data2012-03-16-1331923249.txt'
open(path).readline()
```

```python
import json
record = [json.loads(line) for line in open(path)]
```

```python
tz = [x['tz'] for x in record if 'tz' in x]
# import itertools
# tz

# for name,li in itertools.groupby(tz):
#     print name,list(li)
from collections import defaultdict

def get_counts2(sequence):
    counts = defaultdict(int)
    for x in sequence:
        counts[x] +=1
    return counts
counts = get_counts2(tz)
counts['America/New_York']
arrayToSort = [[y,x] for x,y in counts.items()]
arrayToSort.sort(key= lambda x:x[0])
arrayToSort[-10:]
```

```python
from collections import Counter
c = Counter(tz)
c.most_common(10)
```

```python
from pandas import DataFrame, Series
import pandas as pd
frame = DataFrame(record)
clean_tz = frame['tz'].fillna('Missing')
clean_tz[clean_tz == ''] = 'Unknown'
tz_counts = clean_tz.value_counts()
tz_counts[:10].plot(kind = 'barh', rot=0)
```

```python
results = Series([x.split()[0] for x in frame.a.dropna()])

results[:5]
results.value_counts()[:8]
import numpy as np
cframe = frame[frame.a.notnull()]
operating_system = np.where(cframe['a'].str.contains('Windows'),
                            'Windows', 'Not Windows')
operating_system
```

```python
by_tz_os = cframe.groupby(['tz', operating_system])

agg_counts = by_tz_os.size().unstack().fillna(0)

agg_counts[:10]
```

```python
indexer = agg_counts.sum(1).argsort()
agg_counts.sum(1).iloc[indexer]
res = agg_counts.take(indexer[-10:])
```

```python
res.plot(kind = 'barh')
```

```python
res.div(res.sum(1),axis=0).plot(kind='barh',stacked = True)
```
