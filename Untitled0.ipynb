```python
import pandas as pd
import jieba
import jieba.analyse
import re
import yaha
```

```python
data = pd.read_pickle('data.pickle')
```

```python
with open('text.gbk.txt','wb') as f:
    for body in data.Body:
        if body and len(body)>0:
            f.write((body+'\n').encode('gbk',errors='ignore'))
```

```python
with open('text.utf8.txt','wb') as f:
    for body in data.Body:
        if body and len(body)>0:
            f.write((body+'\n').encode('utf8',errors='ignore'))
```
