```python
def strContain(name, stName):
    return len(set(name) & set(stName)) == len(set(stName))
def nameInRes(name, res):
    stres = None
    for stName in res['name']:
        if strContain(name[0:-6], stName):
            if len(stres) < len(stName)
                stres = stName
    return stres
        
```
