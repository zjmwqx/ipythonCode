```python
year = range(1880,2010)
totalnames = []
for y in year:
    path = '/home/jiminzhou/pydata-book/ch02/names/yob%d.txt'%y
    names = pd.read_csv(path, names=['name','sex','births'])
    names['year']=y
    totalnames.append(names)
```

```python
names = pd.concat(totalnames, ignore_index=True)
```

```python
total_births = names.pivot_table('births', rows='year', cols='sex', aggfunc=sum)
total_births.tail()
total_births.plot()
```

```python
def add_prop(group):
    birth = group.births.astype(float)
    group['prop'] = birth/birth.sum()
    return group
```

```python
def countDiversity(group, q=0.5):
    prop_cumsum = group.sort_index(by='prop',ascending=False).prop.cumsum()
    return np.searchsorted(prop_cumsum.values, 0.5)+1
```

```python
diversity = names.groupby(['year','sex']).apply(countDiversity)
diversity.unstack().plot()
```
