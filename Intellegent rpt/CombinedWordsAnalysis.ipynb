```python
import nltk
bigram_measures = nltk.metrics.BigramAssocMeasures()
def calpmi(entity, dep):
    phraseVal = {}
    resMap = allSentencesInfo[(allSentencesInfo['verb']==dep) & (allSentencesInfo['entity']==entity)]
    if resMap.shape[0] == 0:
        return pd.Series({'entity':None,
                      'dep':None,
                      'mean':None,
                      'var':None,
                      'student_t':None,
                      'chi_sq':None,
                      'pmi':None,
                      'likelihood_ratio':None,})
    phraseVal['mean']=sum(resMap['dist'])/resMap.shape[0]
    phraseVal['var']=resMap['dist'].var()
    phraseVal['freq']=resMap.shape[0]
    n_xx = allSentencesInfo.shape[0]
    n_ii = phraseVal['freq']
    n_xi = wordDF.ix[wordDF['word']==entity]['df'].iloc[0]
    n_ix = wordDF.ix[wordDF['word']==dep]['df'].iloc[0]
    args = (n_ii,(n_ix,n_xi),n_xx)
    return pd.Series({'entity':entity,
                      'dep':dep,
                      'mean':phraseVal['mean'],
                      'var':phraseVal['var'],
                      'student_t':bigram_measures.student_t(*args),
                      'chi_sq':bigram_measures.chi_sq(*args),
                      'pmi':bigram_measures.pmi(*args),
                      'likelihood_ratio':bigram_measures.likelihood_ratio(*args),
})
```
